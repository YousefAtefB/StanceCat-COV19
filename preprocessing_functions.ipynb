```python
import pandas as pd 
import re ,unicodedata
import emoji
import nltk
from nltk.tokenize import word_tokenize
```

```python
df=pd.read_csv("Dataset/dev.csv")
```

```python
def remove_non_ascii(words):
    """Remove non-ASCII characters from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_words.append(new_word)
    return new_words
```

```python
def remove_urls(text):
    return re.sub(r"http\S+", "", text)
```

```python
def remove_flags(text): 
    return re.sub(r"\\p{IsArabic}","h", text)
```

```python
# if it is arabic letter or space 
def get_only_arabic(text):
    val = ""
    for c in text :
        if ((ord(c) in range(1536,1792) or ord(c)==32 or c in emoji.EMOJI_DATA) and ord(c) not in range(1632,1642)):
            val=val+c
            # if char is arabic or sapce or emoji and not number 
        else:
            val=val+" "
    val=re.sub(r" +"," ",val)
    return val
```

```python
print(df["text"][100])
ls=[]
for val in df["text"]: 
    ls.append(get_only_arabic(val))

    
print(ls[100])
```

```python
tokenized_ls=[]
for item in ls : 
    tokenized_ls.append(word_tokenize(item))
```

```python

arb_stopwords = set(nltk.corpus.stopwords.words("arabic"))
for item in tokenized_ls : 
    print(item, len(item))
    for word in item: 
        if word in arb_stopwords:
            item.remove(word)
    print(item, len(item))
```
